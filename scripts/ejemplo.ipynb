```python
import pandas as pd
```

```python
data = pd.read_csv('./libro.csv', sep=';',encoding='utf-8')
data.head(20)
```

```python
def calcular_cuenta_padre(df:pd.DataFrame) -> pd.DataFrame:
	#reseteamos los indices
	df = df.reset_index(drop=True)

	#creamos un id temporal
	df['id_temporal'] = df.index + 1

	#calculamos la longitud del codigo
	df['longitud'] = df['codigo'].apply(lambda x: len(str(x)))
	df['cuenta_padre'] = df['codigo'].apply(lambda x: procesar_cuenta(x, df))
	return df


def procesar_cuenta(x, df):
    longitud = len(str(x))
    if longitud == 2:
        return 0
    else:
        cuenta_padre_code = str(x)[:-1]
		#convertir cuenta padre a integer
        cuenta_padre_code = int(cuenta_padre_code)
        # Convertimos ambos a string para la comparaci√≥n
        mascara = df['codigo'] == cuenta_padre_code
        if mascara.any():
            return df.loc[mascara, 'id_temporal'].iloc[0]
        return 0
```

```python
data = calcular_cuenta_padre(data)
data.head(20)
```
